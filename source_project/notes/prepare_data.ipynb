```python
import json
import os

import pandas as pd
```

```python
data_dir = '/home/niyan/SemEval17-05-kar/data/raw'

mb_train_js = json.load(open(os.path.join(data_dir, 'Microblog_Trainingdata.json'), 'r', encoding='utf-8'))
mb_test_js = json.load(open(os.path.join(data_dir, 'Microblogs_Testdata_withscores.json'), 'r', encoding='utf-8'))
hl_train_js = json.load(open(os.path.join(data_dir, 'Headline_Trainingdata.json'), 'r', encoding='utf-8'))
hl_test_js = json.load(open(os.path.join(data_dir, 'Headlines_Testdata_withscores.json'), 'r', encoding='utf-8'))
```

```python
mb_train_js = [x for x in mb_train_js if len(x['spans']) > 0]
for x in mb_train_js:
    x['spans'] = x['spans'][0]
```

```python
print('Microblog Training Data : {}, Test Data: {}'.format(len(mb_train_js), len(mb_test_js)))
print('Headline Training Data: {}, Test Data: {}'.format(len(hl_train_js), len(hl_test_js)))
```

```python
df_mb_train = pd.DataFrame.from_dict(mb_train_js).rename(columns={'sentiment score': 'sentiment'}).set_index('id')
df_mb_test = pd.DataFrame.from_dict(mb_test_js).rename(columns={'sentiment score': 'sentiment'}).set_index('id')
df_hl_train = pd.DataFrame.from_dict(hl_train_js).set_index('id')
df_hl_test = pd.DataFrame.from_dict(hl_test_js).rename(columns={'sentiment score': 'sentiment'}).set_index('id')
```

```python
df_mb = pd.concat([df_mb_train, df_mb_test]).drop(['text'], axis=1)
df_mb = df_mb.rename(columns={'spans': 'text'})
df_hl = pd.concat([df_hl_train, df_hl_test]).drop(['UniqueID'], axis=1)
df_hl = df_hl.rename(columns={'title': 'text'})
```
